## Carrer Points of the current Drivers

```python
sample1 = data[data['Active'] == True]
sample1
```

```python
plt.figure(figsize=(10,6))
ax1 = sns.barplot(x='Points', y='Driver',data=sample1, hue='Champion')
ax1.set_title("Driver Points")
```

```python
plt.figure(figsize=(10,6))
ax2 =sns.barplot(x='Points_Per_Entry', y='Driver',data=sample1, hue='Champion')
ax2.set_title("Driver Points Per Entry")
```

```python
plt.figure(figsize=(15,10))
ax =sns.scatterplot(x='Win_Rate', y='Podium_Rate',data=sample1, hue='Driver')
ax.set_title("Win Rate vs Podium Rate")
```

```python
plt.figure(figsize=(10,6))
ax3 =sns.barplot(x='FastLap_Rate', y='Driver',data=sample1, hue='Driver')
ax3.set_title("Fastest Lap rate by Drivers")
```
